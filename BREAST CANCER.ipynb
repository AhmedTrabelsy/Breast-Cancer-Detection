```python
import os, shutil
from sklearn.model_selection import train_test_split

# ğŸ“ Dossier racine
root_dir = r"C:\Users\VICTUS\Documents\Augmented Dataset"

# ğŸ“ Dossiers de sortie
train_dir = "datasett/train"
val_dir = "datasett/val"

classes = ["Cancer", "Non-Cancer"]

# ğŸ”¹ CrÃ©ation des dossiers
for cls in classes:
    os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
    os.makedirs(os.path.join(val_dir, cls), exist_ok=True)

# ğŸ”¹ Collecte des images
cancer_images = [
    os.path.join(root_dir, "Cancer", f)
    for f in os.listdir(os.path.join(root_dir, "Cancer"))
    if f.lower().endswith((".png", ".jpg", ".jpeg"))
]

non_cancer_images = [
    os.path.join(root_dir, "Non-Cancer", f)
    for f in os.listdir(os.path.join(root_dir, "Non-Cancer"))
    if f.lower().endswith((".png", ".jpg", ".jpeg"))
]

print(f"ğŸ” Images trouvÃ©es : {len(cancer_images)} Cancer | {len(non_cancer_images)} Non-Cancer")

# ğŸ›‘ SÃ©curitÃ©
if len(cancer_images) == 0 or len(non_cancer_images) == 0:
    raise ValueError("âŒ Une classe est vide")

# ğŸ”¹ Split 80/20
train_c, val_c = train_test_split(cancer_images, test_size=0.2, random_state=42)
train_nc, val_nc = train_test_split(non_cancer_images, test_size=0.2, random_state=42)

def copy_images(images, target):
    for img in images:
        shutil.copy(img, target)

# ğŸ”¹ Copie
copy_images(train_c, os.path.join(train_dir, "Cancer"))
copy_images(val_c, os.path.join(val_dir, "Cancer"))

copy_images(train_nc, os.path.join(train_dir, "Non-Cancer"))
copy_images(val_nc, os.path.join(val_dir, "Non-Cancer"))

print("âœ… Dataset prÃªt avec succÃ¨s !")
```
