```python
import os, shutil
from sklearn.model_selection import train_test_split

# ğŸ“ Ton dossier d'archive dÃ©compressÃ©
root_dir = r"C:\Users\VICTUS\Downloads\archive"

# ğŸ“ Dossiers de sortie
train_dir = "dataset/train"
val_dir = "dataset/val"

# ğŸ”¹ CrÃ©ation des sous-dossiers sâ€™ils nâ€™existent pas
for cls in ["benign", "malignant"]:
    os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
    os.makedirs(os.path.join(val_dir, cls), exist_ok=True)

# ğŸ”¹ Collecte des chemins dâ€™images
benign_images, malignant_images = [], []

for folder in os.listdir(root_dir):
    sub_path = os.path.join(root_dir, folder)
    if not os.path.isdir(sub_path):
        continue

    benign_path = os.path.join(sub_path, "0")
    malignant_path = os.path.join(sub_path, "1")

    if os.path.exists(benign_path):
        benign_images += [os.path.join(benign_path, f) for f in os.listdir(benign_path) if f.endswith((".png", ".jpg", ".jpeg"))]
    if os.path.exists(malignant_path):
        malignant_images += [os.path.join(malignant_path, f) for f in os.listdir(malignant_path) if f.endswith((".png", ".jpg", ".jpeg"))]

print(f"ğŸ” Images trouvÃ©es : {len(benign_images)} bÃ©nignes et {len(malignant_images)} malignes")

# ğŸ”¹ Division train/val (80/20)
train_benign, val_benign = train_test_split(benign_images, test_size=0.2, random_state=42)
train_malignant, val_malignant = train_test_split(malignant_images, test_size=0.2, random_state=42)

def copy_images(img_list, target_folder):
    for img_path in img_list:
        try:
            shutil.copy(img_path, target_folder)
        except Exception as e:
            pass

print("ğŸ“¦ Copie des images d'entraÃ®nement...")
copy_images(train_benign, os.path.join(train_dir, "benign"))
copy_images(train_malignant, os.path.join(train_dir, "malignant"))

print("ğŸ“¦ Copie des images de validation...")
copy_images(val_benign, os.path.join(val_dir, "benign"))
copy_images(val_malignant, os.path.join(val_dir, "malignant"))

print("âœ… Dataset prÃªt ! Structure crÃ©Ã©e avec succÃ¨s.")
```
